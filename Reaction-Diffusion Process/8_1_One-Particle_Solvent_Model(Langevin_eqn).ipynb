## Overview
---
The force on the particle, $F(t),$ is expressed as the sum of the frictional force $F_{f}(t)=-\gamma V(t)$ and a random force, namely as:

$$
F(t)=-\gamma V(t)+\gamma \sqrt{2 D}\frac{\mathrm{d} W}{\mathrm{d} t}\text{---(1)}
$$

This equation is also called **Langevin equation**. To get some initial insights into the problem of solvent molecule behaviour, we restrict to a one-dimensional model where the postion $X(t)$ and velocity $V(t)$ of the diffusing particle evolve according to 

$$
\begin{aligned} \frac{\mathrm{d} X}{\mathrm{d} t}(t) &=V(t) \\ m_{p} \frac{\mathrm{d} V}{\mathrm{d} t}(t) &=F(t) \end{aligned}\text{---(2)}
$$

We denote the positions and velocities of the solvent molecules by the lower-case letters $x_i(t)$ and $v_i(t)$, respectively, where $i=1,2,...,N$. Now, we simplify the model by assuming $N=1$.

Force term $F(t)$ is no longer given by (1) but by interaction with the "solvent molecule". We assume that this interaction is given as a harmonic spring with spring constant $k,$ i.e. $F(t)=k(x(t)-X(t))$. Then equations in $(2)$ read as follows:

$$
\begin{aligned} \frac{\mathrm{d} \mathbf{X}}{\mathrm{d} t}(t) &=V(t)\text{---(3)} \\ m_{p} \frac{\mathrm{d} V}{\mathrm{d} t}(t) &=k(x(t)-X(t))\text{---(4)} \end{aligned}
$$

Equations (3) are coupled with the evolution equations for the "solvent molecule" of mass $m_s$:

$$
\begin{aligned} \frac{\mathrm{d} x}{\mathrm{d} t}(t) &=v(t)\text{---(5)} \\ m_{s} \frac{\mathrm{d} v}{\mathrm{d} t}(t) &=k(X(t)-x(t))-\gamma_{s} v(t)+\gamma_{s} \sqrt{2 D_{s}}\frac{dW}{dt}\text{---(6)} \end{aligned}
$$

The equation system(3)-(6) is so-called the **one-particle solvent model.**

While equation (3)-(5) are mathematically well-defined, the equation (6) needs more consideration: 

>From the physical point of view, the friction and random terms in (6) implicitly take into account that there are many other solvent molecules in the system which interact with our one explicitly described "solvent molecule".

> We also discussed through Einstein–Smoluchowski relation that $D_s$ in (6) is related to temperature. Consequently, the friction and random terms in (6) can be viewed as a "stochastic thermostat",**keeping the simulated system at the right temperature.**

> One could also argue that the friction and noise terms should not only be in (6), but also on the RHS of eqn.(4) as other solvent molecules that implicitly interact with
the “solvent molecule” at position $x(t)$ would also interact with the diffusing particle at position $X(t)$. In the case of one-solvent particle model, we ignore those terms for simplification.

## Generalized Langevin Equation
---
Our next goal is to derive a system of equations for the important variables $X$ and $V .$ This will be given in the form of the so-called generalized Langevin
equation. 

If we rewrite the equations (3)-(6) by using four variables $X(t),V(t),U(t)$ and $Z(t)$, where 
$$
U(t)=\frac{k(x(t)-X(t))}{m_{p}} \quad \text { and } \quad Z(t)=\frac{k v(t)}{m_{p}}
$$
we have
$$
\begin{aligned} X(t+\mathrm{d} t) &=X(t)+V(t) \mathrm{d} t\text{---(7)} \\ V(t+\mathrm{d} t) &=V(t)+U(t) \mathrm{d} t\text{---(8)} \\ U(t+\mathrm{d} t) &=U(t)+\left(Z(t)-\alpha_{1} V(t)\right) \mathrm{d} t\text{---(9)} \\ Z(t+\mathrm{d} t) &=Z(t)-\alpha_{4}\left(Z(t)+\alpha_{2} U(t)\right) \mathrm{d} t+\alpha_{3} \alpha_{4} \mathrm{d} W \text{---(10)}\end{aligned}
$$

where U is the acceleration of the diffusing particle, and $\alpha_j$ are positive parameters given by:
$$
\alpha_{1}=\frac{k}{m_{p}}, \quad \alpha_{2}=\frac{k}{\gamma_{s}}, \quad \alpha_{3}=\frac{k \sqrt{2 D_{s}}}{m_{p}} \quad \text { and } \quad \alpha_{4}=\frac{\gamma_{s}}{m_{s}}\text{---(11)}
$$

**We first consider equations (7)-(10) under the limit of $\alpha_{4} \rightarrow \infty$**. Dividing equation (10) by $\alpha_{4}$ and passing to the limit:

$$
Z(t) \mathrm{d} t \approx-\alpha_{2} U(t) \mathrm{d} t+\alpha_{3} \mathrm{d} W
$$

Substituting for Z in equation (9), we reduce the one-particle solvent model into:

$$
\begin{aligned} X(t+\mathrm{d} t) &=X(t)+V(t) \mathrm{d} t\text{--(12)} \\ V(t+\mathrm{d} t) &=V(t)+U(t) \mathrm{d} t\text{--(13)} \\ U(t+\mathrm{d} t) &=U(t)-\left(\alpha_{1} V(t)+\alpha_{2} U(t)\right) \mathrm{d} t+\alpha_{3} \mathrm{d} W\text{--(14)} \end{aligned}
$$

Now we use the computational definition to interpret equation (14) as:

$$
U(t+\Delta t)=\left(1-\alpha_{2} \Delta t\right) U(t)-\alpha_{1} V(t) \Delta t+\alpha_{3} \sqrt{\Delta t} \xi\text{--(15)} 
$$

In a computer simulation, we use equation (15) iteratively at times $t=0, \Delta t$ $2 \Delta t, \ldots,$ namely we use the following $k$ equations to compute the value of $U$ at time $k \Delta t,$ for $k=1,2,3, \ldots$ :

$$
\begin{aligned} U(\Delta t) &=\left(1-\alpha_{2} \Delta t\right) U(0)-\alpha_{1} V(0) \Delta t+\alpha_{3} \sqrt{\Delta t} \xi_{0} \\ U(2 \Delta t) &=\left(1-\alpha_{2} \Delta t\right) U(\Delta t)-\alpha_{1} V(\Delta t) \Delta t+\alpha_{3} \sqrt{\Delta t} \xi_{1} \\ U(3 \Delta t) &=\left(1-\alpha_{2} \Delta t\right) U(2 \Delta t)-\alpha_{1} V(2 \Delta t) \Delta t+\alpha_{3} \sqrt{\Delta t} \xi_{2} \\ & \vdots \\ U(k \Delta t) &=\left(1-\alpha_{2} \Delta t\right) U((k-1) \Delta t)-\alpha_{1} V((k-1) \Delta t) \Delta t+\alpha_{3} \sqrt{\Delta t} \xi_{k-1} \end{aligned}
$$

where $\xi_{1}, \xi_{2}, \xi_{3}, \ldots, \xi_{k-1}$ are random numbers which are independently
sampled from the normal distribution with zero mean and unit variance.

**Substituting these equations iteratively from the first one to the last one, we have**:

$$
\begin{aligned} U(k \Delta t)=&\left(1-\alpha_{2} \Delta t\right)^{k} U(0)-\alpha_{1} \Delta t \sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{i-1} V((k-i) \Delta t) \\ &+\alpha_{3} \sqrt{\Delta t} \sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{i-1} \xi_{k-i} \end{aligned}\text{---(16)}
$$

**since $\left(1-\alpha_{2} \Delta t\right)<1$, the first term on the right-hand side will become smaller and smaller as we increase k. Put another way, as time progresses, the initial value of the acceleration will be forgotten by the system.** considering sufficiently long times $t = k\Delta t$, we neglect the first term on the right-hand side of (16).

The last term on the right-hand side is the sum of independent normally distributed random variables, i.e. **it is also normally distributed with variance equal to the sum of the individual variances, which can be approximated by**(Using $Var(X) = E(X^2)-E^2(X)$):

$$
\alpha_{3}^{2} \Delta t \sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{2 i-2}=\alpha_{3}^{2} \Delta t \frac{\left(1-\alpha_{2} \Delta t\right)^{2 k}-1}{\left(1-\alpha_{2} \Delta t\right)^{2}-1} \approx \frac{\alpha_{3}^{2} \Delta t}{1-\left(1-\alpha_{2} \Delta t\right)^{2}} \approx \frac{\alpha_{3}^{2}}{2 \alpha_{2}}
$$

where, in the first approximation, we have again used the fact that $k$ is sufficiently large so that we can neglect $\left(1-\alpha_{2} \Delta t\right)^{2 k},$ while in the second approximation, we used the fact that $\Delta t$ is small (so that $\alpha_{2} \Delta t \ll 1$ ).

Consequently, equation(16) becomes:

$$
U(t)=-\alpha_{1} \Delta t \sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{i-1} V((k-i) \Delta t)+\frac{\alpha_{3}}{\sqrt{2 \alpha_{2}}} \xi(t)\text{---(17)}
$$

where $t=k \Delta t$ is (sufficiently large) time and $\xi(t)$ is a random number which is sampled from the normal distribution with zero mean and unit variance.**Using approximation $\left(1-\alpha_{2} \Delta t\right)^{i} \approx \exp \left(-i \alpha_{2} \Delta t\right),$ which is valid for sufficiently small $\Delta t\left(\text { so that }\left(i \alpha_{2} \Delta t\right)^{2} \ll 1\right),$ we obtain** 

$$
\begin{aligned} \Delta t \sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{i-1} V((k-i) \Delta t) & \approx \sum_{i=1}^{k} \exp \left(-\alpha_{2}(i-1) \Delta t\right) V((k-i) \Delta t) \Delta t \\ & \approx \int_{0}^{t} \exp \left(-\alpha_{2} \tau\right) V(t-\tau) \mathrm{d} \tau \end{aligned}
$$

where the last approximation is valid in the limit $\Delta t \rightarrow 0 .$ And (17) is simplified as:

$$
U(t)=-\alpha_{1} \int_{0}^{t} \exp \left(-\alpha_{2} \tau\right) V(t-\tau) \mathrm{d} \tau+\frac{\alpha_{3}}{\sqrt{2 \alpha_{2}}} \xi(t)\text{---(18)}
$$

>**We define the friction kernel $\kappa(\tau)$ for $\tau \geq 0$ and noise term $R(t)$ for $t \geq 0$ by:**
$$
\begin{aligned} \kappa(\tau) &=\alpha_{1} \exp \left(-\alpha_{2} \tau\right) \\ R(t) &=\frac{\alpha_{3}}{\sqrt{2 \alpha_{2}}} \xi(t) \end{aligned}
$$

Substituting (18) back into (8) and (9), we obtain **the generalized Langevien equation in the form:**
>$$
\begin{aligned} \frac{\mathrm{d} X}{\mathrm{d} t} &=V(t)\text{---(19)} \\ \frac{\mathrm{d} V}{\mathrm{d} t} &=-\int_{0}^{t} \kappa(\tau) V(t-\tau) \mathrm{d} \tau+R(t)\text{---(20)} \end{aligned}
$$

## Generalized Fluctuation–Dissipation theorem
---
Unlike in the case of the standard Langevin equation
with force term (1), **we have no mathematical issues in writing the generalized Langevin equation as a system of differential equations, because the
noise term R(t) is a continuous function**. **It depends on the noise terms at previous times, which we can formally quantify by computing its autocorrelation function**

$$
\langle R(t) R(t-\tau)\rangle=\mathrm{E}[R(t) R(t-\tau)]=\alpha_{3}^{2} \Delta t\left\langle\sum_{i=1}^{k}\left(1-\alpha_{2} \Delta t\right)^{i-1} \xi_{k-i} \sum_{\ell=1}^{k-j}\left(1-\alpha_{2} \Delta t\right)^{\ell-1} \xi_{k-j-\ell}\right\rangle
$$

Using $E\left[\xi_{i} \xi_{j}\right]=0$ for $i \neq j$ and $E\left[\xi_{i}^{2}\right]=1,$ we have

$$
\langle R(t) R(t-\tau)\rangle=\alpha_{3}^{2} \Delta t \sum_{\ell=1}^{k-j}\left(1-\alpha_{2} \Delta t\right)^{j+2 \ell-2}
$$

Considering that $k$ can be as large as we want (while $j$ is fixed), we can approximate

$$
\langle R(t) R(t-\tau)\rangle \approx \alpha_{3}^{2} \Delta t\left(1-\alpha_{2} \Delta t\right)^{j} \sum_{\ell=0}^{\infty}\left(1-\alpha_{2} \Delta t\right)^{2 \ell}=\frac{\alpha_{3}^{2} \Delta t\left(1-\alpha_{2} \Delta t\right)^{j}}{1-\left(1-\alpha_{2} \Delta t\right)^{2}}
$$

Finally, using that $\alpha_{2} \Delta t \ll 1,$ we have

$$
\langle R(t) R(t-\tau)\rangle \approx \frac{\alpha_{3}^{2} \exp \left(-\alpha_{2} \tau\right)}{2 \alpha_{2}}\text{---(21)}
$$

Using the definition of $\kappa(t)$, the (21) can be rewritten as the **generalized fluctuation-dissipation theorem:**:

$$
\langle R(t) R(t-\tau)\rangle=\frac{\alpha_{3}^{2}}{2 \alpha_{1} \alpha_{2}} \kappa(\tau)\text{---(22)}
$$

## Loose the Limit of $\alpha_4\rightarrow \infty$
---
If we generalize the results above to the full one-particle solvent model expressed in terms of equations (7)-(10), the generalized Langevin equation (19)-(20) holds with the definition of the friction kernel:
$$
\kappa(\tau)=\alpha_{1} \exp \left(-\frac{\alpha_{4} \tau}{2}\right)\left(\frac{\alpha_{4}}{\alpha_{5}} \sinh \left(\frac{\alpha_{5} \tau}{2}\right)+\cosh \left(\frac{\alpha_{5} \tau}{2}\right)\right)\text{---(23)}
$$
where $\alpha_{5}=\sqrt{\alpha_{4}\left(\alpha_{4}-4 \alpha_{2}\right)}$.

> The auxiliary coefficient $\alpha_{5}$ is a square root of a real negative number for $\alpha_{4}<4 \alpha_{2}$. However formula (23) is still valid in this case - for $\alpha_{4}<4 \alpha_{2}$ it can be rewritten in terms of sine and cosine functions, taking into account that $\alpha_{5}$ is purely imaginary $i|\alpha_5|$ and using identities: $\sinh \left(\mathbf{i}\left|\alpha_{5}\right|\right)=\mathrm{i} \sin \left(\left|\alpha_{5}\right|\right)$ and $\cosh \left(\mathbf{i}\left|\alpha_{5}\right|\right)=\cos \left(\left|\alpha_{5}\right|\right)$
