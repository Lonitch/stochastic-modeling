The phenomenon we present is sometimes called self-induced stochastic resonance (Muratov et al., 2005; DeVille et al., 2006). We consider a system of two chemical species $A$ and $B$ in a reactor of volume $v$ which are subject to the chemical reactions:
$$
2 A+B \stackrel{k_{1}}{\longrightarrow} 3 A
$$
$$
\emptyset \overset{k2}{\underset{k_{3}}\leftrightarrow} A
$$
$$
\emptyset \stackrel{k_{4}}{\longrightarrow} B
$$
The approximate deterministic description is given by the system of ODEs
$$
\begin{aligned} \frac{\mathrm{d} a}{\mathrm{d} t} &=k_{1} a^{2} b+k_{2}-k_{3} a \\ \frac{\mathrm{d} b}{\mathrm{d} t} &=-k_{1} a^{2} b+k_{4} \end{aligned}
$$
with $\bar{A}(t)=a(t) v$ and $\bar{B}(t)=b(t) v$,
$$
\begin{aligned} \frac{\mathrm{d} \bar{A}}{\mathrm{d} t} &=\frac{k_{1}}{v^{2}} \bar{A}^{2} \bar{B}+k_{2} v-k_{3} \bar{A} \\ \frac{\mathrm{d} \bar{B}}{\mathrm{d} t} &=-\frac{k_{1}}{v^{2}} \bar{A}^{2} \bar{B}+k_{4} v \end{aligned}
$$

## Bifurcation of ODE system
The following implementation shows that the behavior of ODE system introduced above changes dramatically when $k_4v$ is varied from 25 $sec^{-1}$ to 100 $sec^{-1}$. **There exists a unique constant $k_{b}$ (bifurcation value) in the interval $(25,100)$ sec $^{-1}$ such that the solution of the ODEs converges to the steady state for $k_{4} v<k_{b}$ and oscillates for $k_{4} v>k_{b} .$** The value of the constant $k_{b}$ can be identified by investigating the stability of the steady state of the ODEs.

```python
import numpy as np
from numpy.random import uniform
from scipy.integrate import odeint

def resonance(k1,k2,k3,k4,v,T,A0=10,B0=10):
    a4 = k4*v
    a2 = k2*v
    tlst = [0]
    ct={'A':[A0],'B':[B0]}
    responds={0:[+1,-1],1:[1,0],2:[-1,0],3:[0,1]}
    while tlst[-1]<T:
        a,b = ct['A'][-1],ct['B'][-1]
        if a>1:
            a1 = a*(a-1)/2*b*k1/v**2
        else:
            a1=0
        a3=k3*a
        a0 = a1+a2+a3+a4
        plst = np.cumsum([a1,a2,a3,a4])/a0
        r1,r2 = uniform(size=2)
        tau = 1/a0*np.log(1/r1)
        tlst.append(tlst[-1]+tau)
        dA=responds[sum(plst<r2)]
        ct['A'].append(ct['A'][-1]+dA[0])
        ct['B'].append(ct['B'][-1]+dA[1])
    return tlst,ct

def solvODE(k1,k2,k3,k4,v,T,A0=10,B0=10,nstep=100000):
    p1,p2,p3,p4=k1/v**2,k2*v,k3,k4*v
    
    def func(y,t,f1,f2,f3,f4):
        a,b = y
        return [f1*a**2*b+f2-f3*a, -f1*a**2*b+f4]
    
    tt = np.linspace(0, T, nstep)
    sol = odeint(func, [A0,B0], tt, args=(p1,p2,p3,p4))
    return tt,sol
```

```python
V=1
K1=0.00004
K2=50
K3=10
K4=25
T=8000
simt,simc=resonance(K1,K2,K3,K4,V,T)
odet,odec=solvODE(K1,K2,K3,K4,V,T)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
simA = np.array(simc['A'])+1 # remove 0
plt.plot(simt,np.log10(simA),'b-',label='Simulation')
plt.plot(odet,np.log10(odec[:,0]),'r-',linewidth=2,label='ODE')
plt.xlabel('Time[sec]')
plt.ylabel('log10(A(t))')
plt.legend(loc='upper left')
```

```python
simB = np.array(simc['B'])+1 # remove 0
plt.plot(simt,simB,'b-',label='Simulation')
plt.plot(odet,odec[:,1],'r-',linewidth=2,label='ODE')
plt.xlabel('Time[sec]')
plt.ylabel('B(t)')
plt.legend(loc='upper left')
```

**Now, let's try to change $k_4$**

```python
K4=60
T=2000
simt,simc=resonance(K1,K2,K3,K4,V,T)
odet,odec=solvODE(K1,K2,K3,K4,V,T)
simA = np.array(simc['A'])+1 # remove 0
plt.plot(simt,np.log10(simA),'b-',label='Simulation')
plt.plot(odet,np.log10(odec[:,0]),'r-',linewidth=2,label='ODE')
plt.xlabel('Time[sec]')
plt.ylabel('log10(A(t))')
plt.legend(loc='upper left')
```

```python
simB = np.array(simc['B'])+1 # remove 0
plt.plot(simt,simB,'b-',label='Simulation')
plt.plot(odet,odec[:,1],'r-',linewidth=2,label='ODE')
plt.xlabel('Time[sec]')
plt.ylabel('B(t)')
plt.legend(loc='upper left')
```

```python
plt.plot(np.log10(simA),simB,'-',color='skyblue',label='Simulation')
plt.plot(np.log10(odec[:,0]),odec[:,1],'r-',linewidth=2,label='ODE')
temp=np.linspace(1,1000,10000)
plt.plot(np.log10(temp),(K3*temp-K2*V)/(temp**2*K1/V**2),'g-',label='[dA/dt=0] nullcline')
plt.plot(np.log10(temp),(K4*V)/(temp**2*K1/V**2),'g--',label='[dB/dt=0] nullcline')
plt.xlabel('log10(A(t))')
plt.ylabel('B(t)')
plt.ylim((0,20000))
plt.legend(loc='upper right')
```

The $[\mathrm{d} \bar{A} / \mathrm{d} t=0]$ nullcline is given by
$$
\bar{B}=\frac{k_{3} \bar{A}-k_{2} v}{\bar{A}^{2} k_{1} / v^{2}}
$$
The maximum of this curve is
$$
a_{m}=\frac{2 k_{2} v}{k_{3}}=10
$$
and it is equal to $b_{m}=12500$.
The $[\mathrm{d} \bar{B} / \mathrm{d} t=0]$ nullcline is given by
$$
\bar{B}=\frac{k_{4} v}{\bar{A}^{2} k_{1} / v^{2}}
$$
**Thus the bifurcation value $k_{b}$ can be approximated as**
$$
k_{b} \doteq a_{m}^{2} b_{m} \frac{k_{1}}{v^{2}}=50 \sec ^{-1}
$$
**In the plot above, the two nullcline intersects at the right branch, bringing oscillation in the ODE system.** The following plot has nullcline intersection at the left branch ($k_4=40$), and no oscillation would happen.

```python
K4=40
T=2000
simt,simc=resonance(K1,K2,K3,K4,V,T)
odet,odec=solvODE(K1,K2,K3,K4,V,T)
plt.plot(np.log10(simA),simB,'-',color='skyblue',label='Simulation')
plt.plot(np.log10(odec[:,0]),odec[:,1],'r-',linewidth=5,label='ODE')
temp=np.linspace(1,1000,10000)
plt.plot(np.log10(temp),(K3*temp-K2*V)/(temp**2*K1/V**2),'g-',label='[dA/dt=0] nullcline')
plt.plot(np.log10(temp),(K4*V)/(temp**2*K1/V**2),'g--',label='[dB/dt=0] nullcline')
plt.xlabel('log10(A(t))')
plt.ylabel('B(t)')
plt.ylim((0,20000))
plt.legend(loc='upper right')
```
