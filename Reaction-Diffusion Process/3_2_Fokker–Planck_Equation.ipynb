## Overview

Let us suppose that $X(t)$ evolves according to the following SDE:
$$
X(t+\mathrm{d} t)=X(t)+f(X(t), t) \mathrm{d} t+g(X(t), t) \mathrm{d} W\text{---(1)}
$$
We define the **probability distribution function $p(x, t)$ so that $p(x, t)$ d $x$ is the probability that
$X(t) \in[x, x+\mathrm{d} x]$**. Since $X(t)$ must be somewhere on the real line, the function $p(x, t)$ satisfies the normalization condition:
$$
\int_{\mathbb{R}} p(x, t) \mathrm{d} x=1\text{---(2)}
$$
Note that $p(x,t)$ can be calculated by the SSA introduced in the notebook (3.1). But we can calculate it more efficiently without performing any stochastic simulation. We know that the $p(x,t)$ evolves according to the **Fokker–Planck equation**:

$$
\frac{\partial p}{\partial t}(x, t)=\frac{\partial^{2}}{\partial x^{2}}\left(\frac{g^{2}(x, t)}{2} p(x, t)\right)-\frac{\partial}{\partial x}(f(x, t) p(x, t))\text{---(3)}
$$

This PDE is also called as *Kolmogorov forward equation or the Smoluchowski equation*.

## Applications of Fokker-Planck equation

If we have $f(x, t) \equiv 0$ and $g(x, t) \equiv 1$, Fokker-Planck equation becomes:

$$
\frac{\partial p}{\partial t}(x, t)=\frac{1}{2} \frac{\partial^{2} p}{\partial x^{2}}(x, t)\text{---(3)}
$$

We imposed the initial condition $X(0)=0$ for the SDE, and we have IC as $p(x, 0)=\delta(x)$. The solution is therefore:

$$
p(x, t)=\frac{1}{\sqrt{2 \pi t}} \exp \left[-\frac{x^{2}}{2 t}\right]
$$

Suppose now that the coefficients $f$ and $g$ of the $\mathrm{SDE}(2)$ are arbitrary functions of $x,$ but do not depend on time, i.e. $f(x, t) \equiv f(x)$ and $g(x, t) \equiv$$g(x) .$ The long-time behavior of such SDEs is reasonably characterized by the stationary distribution:

$$
p_{S}(x)=\lim _{t \rightarrow \infty} p(x, t)
$$

By solving the following equation:

$$
\frac{\mathrm{d}^{2}}{\mathrm{d} x^{2}}\left(\frac{g^{2}(x)}{2} p_{s}(x)\right)-\frac{\mathrm{d}}{\mathrm{d} x}\left(f(x) p_{s}(x)\right)=0
$$

we obtain:

$$
p_{s}(x)=\frac{C}{g^{2}(x)} \exp \left[\int_{0}^{x} \frac{2 f(y)}{g^{2}(y)} \mathrm{d} y\right]
$$
Because $\int_{\mathbb{R}} p_{s}(x) \mathrm{d} x=1$, we have the constant $C$ as:

$$
C=\left(\int_{\mathbb{R}} \frac{1}{g^{2}(x)} \exp \left[\int_{0}^{x} \frac{2 f(y)}{g^{2}(y)} \mathrm{d} y\right] \mathrm{d} x\right)^{-1}
$$

## Chapman–Kolmogorov equation

We now let $p(x, t | y, s)$ d $x$ be the probability that $X(t) \in[x, x+\mathrm{d} x)$ under the condition that $X(s)=y,$ where $s$ and $t$ are arbitrary time points with $s<t .$

Now consider the value of $X$ at time $t+\Delta t .$ The probability that the variable $X$ is in the interval $[z, z+\mathrm{d} z)$ at time $t+\Delta t(\text { i.e. } X(t+\Delta t) \in[z, z+\mathrm{d} z))$ is equal to $p(z, t+\Delta t | y, s) \mathrm{d} z .$ We now divide this time interval from $s$ to $t+\Delta t$ into a time interval from $s$ to $t,$ and an interval from $t$ to $t+\Delta t$. At time $t$ the variable $X$ takes on a value in the interval $[x, x+\mathrm{d} x)$ with probability $p(x, t | y, s) \mathrm{d} x$. It then moves from $[x, x+\mathrm{d} x)$ to the interval $[z, z+\mathrm{d} z)$ over the time interval $t$ to $t+\Delta t$ with probability $p(z, t+\Delta t | x, t) d z .$ To calculate the probability that $X$ moves from the value $y$ at time $s$ to $[z, z+d z)$ at time $t+\Delta t,$ we need to **integrate over all possible intermediate points $x .$ In doing so we obtain the so-called Chapman-Kolmogorov equation**

$$
p(z, t+\Delta t | y, s)=\int_{\mathbb{R}} p(z, t+\Delta t | x, t) p(x, t | y, s) \mathrm{d} x
$$

where $s<t$. **This equation is valid for all $\Delta t$, not just small $\Delta t$**.

**To derive Fokker-Planck equation we now take the limit as $\Delta t \rightarrow 0 .$ We first multiply both sides by a smooth test function $\varphi(z)$ and integrate over $z,$ to give**

$$
\int_{\mathbb{R}} p(z, t+\Delta t | y, s) \varphi(z) \mathrm{d} z=\int_{\mathbb{R}}\left[\int_{\mathbb{R}} p(z, t+\Delta t | x, t) \varphi(z) \mathrm{d} z\right] p(x, t | y, s) \mathrm{d} x
$$

We rename the integration variable $z$ to be $x$ on the left-hand side to get

$$
\int_{\mathbb{R}} p(x, t+\Delta t | y, s) \varphi(x) \mathrm{d} x=\int_{\mathbb{R}}\left[\int_{\mathbb{R}} p(z, t+\Delta t | x, t) \varphi(z) \mathrm{d} z\right] p(x, t | y, s) \mathrm{d} x
$$

Now using the Taylor expansion of $\varphi(z)$ around the point $x$ on the right-hand side, we get

$$
\begin{array}{l}{\int_{\mathbb{R}} p(x, t+\Delta t | x, s) \varphi(x) \mathrm{d} x} \\ {\quad=\int_{\mathbb{R}}\left[\int_{\mathbb{R}} p(z, t+\Delta t | x, t)\left(\varphi(x)+\varphi^{\prime}(x)(z-x)\right.\right.} \\ {\left.\left.\quad+\varphi^{\prime \prime}(x) \frac{(z-x)^{2}}{2}+o\left((z-x)^{2}\right)\right) \mathrm{d} z\right] p(x, t | y, s) \mathrm{d} x}\end{array}
$$
$$
\begin{array}{l}{=\int_{\mathbb{R}}\left[\varphi(x) \int_{\mathbb{R}} p(z, t+\Delta t | x, t) \mathrm{d} z+\varphi^{\prime}(x) \int_{\mathbb{R}}(z-x) p(z, t+\Delta t | x, t) \mathrm{d} z\right.} \\ {\quad+\frac{1}{2} \varphi^{\prime \prime}(x) \int_{\mathbb{R}}(z-x)^{2} p(z, t+\Delta t | x, t) \mathrm{d} z} \\ {\left.\quad+\int_{\mathbb{R}} o\left((z-x)^{2}\right) p(z, t+\Delta t | x, t) \mathrm{d} z\right] p(x, t | y, s) \mathrm{d} x}\end{array}\text{---(4)}
$$

Next, we simplify the integrals on the right-hand side. Starting at $X(t)=x,$ the probability that the trajectory arrives somewhere at time $t+\Delta t$ is equal to 1 (we
do not lose the trajectory). Thus

$$
\int_{\mathbb{R}} p(z, t+\Delta t | x, t) \mathrm{d} z=1\text{--(5)}
$$
$$
\mathrm{E}[X(t+\Delta t)-x | X(t)=x]=\int_{\mathbb{R}}(z-x) p(z, t+\Delta t | x, t) \mathrm{d} z
$$

Using eqn.(1). Thus, 

$$
\int_{\mathbb{R}}(z-x) p(z, t+\Delta t | x, t) \mathrm{d} z=f(x, t) \Delta t\text{---(6)}
$$

Again, using eqn. (1), we obtain:

$$
\mathrm{E}\left[(X(t+\Delta t)-x)^{2} | X(t)=x\right]=\int_{\mathbb{R}}(z-x)^{2} p(z, t+\Delta t | x, t) \mathrm{d} z
$$

$$
\begin{array}{l}{\mathrm{E}\left[(X(t+\Delta t)-x)^{2} | X(t)=x\right]=\mathrm{E}\left[(f(x, t) \Delta t+g(x, t) \sqrt{\Delta t} \xi)^{2}\right]} \\ {\quad=f(x, t)^{2} \Delta t^{2}+2 f(x, t) g(x, t) \Delta t^{3 / 2} \mathrm{E}[\xi]+g(x, t)^{2} \Delta t \mathrm{E}\left[\xi^{2}\right]} \\ {\quad=g(x, t)^{2} \Delta t+O\left(\Delta t^{2}\right)}\end{array}
$$

Consequently,

$$
\int_{\mathbb{R}}(z-x)^{2} p(z, t+\Delta t | x, t) \mathrm{d} z=g(x, t)^{2} \Delta t+O\left(\Delta t^{2}\right)\text{---(7)}
$$

Substitute (2),(5),(6),(7) into (4), we have:

$$
\begin{array}{l}{\int_{\mathbb{R}} \frac{p(x, t+\Delta t | y, s)-p(x, t | y, s)}{\Delta t} \varphi(x) \mathrm{d} x} \\ {\quad=\int_{\mathbb{R}} \varphi^{\prime}(x) f(x, t) p(x, t | y, s) \mathrm{d} x+\int_{\mathbb{R}} \varphi^{\prime \prime}(x) \frac{g(x, t)^{2}}{2} p(x, t | y, s) \mathrm{d} x+O(\Delta t)}\end{array}
$$

Using integration by parts on the right-hand side, we get

$$
\begin{aligned} 0=& \int_{\mathbb{R}} \varphi(x) \times\left\{-\frac{p(x, t+\Delta t | y, s)-p(x, t | y, s)}{\Delta t}\right.\\ &\left.-\frac{\partial}{\partial x}(f(x, t) p(x, t | y, s))+\frac{\partial^{2}}{\partial x^{2}}\left(\frac{g(x, t)^{2}}{2} p(x, t | y, s)\right)\right\} \mathrm{d} x+O(\Delta t) \end{aligned}
$$

**since the test function $\varphi(x)$ is arbitrary, we conclude that the term inside curly brackets must be zero. Thus**

$$
\begin{aligned} \frac{p(x, t+\Delta t | y, s)-p(x, t | y, s)}{\Delta t} & \\=\frac{\partial^{2}}{\partial x^{2}}\left(\frac{g(x, t)^{2}}{2} p(x, t | y, s)\right)-\frac{\partial}{\partial x}(f(x, t) p(x, t | y, s))+O(\Delta t) \end{aligned}
$$

Passing to the limit $\Delta t \rightarrow 0,$ we obtain the Fokker-Planck equation in the form

$$
\frac{\partial}{\partial t} p(x, t | y, s)=\frac{\partial^{2}}{\partial x^{2}}\left(\frac{g(x, t)^{2}}{2} p(x, t | y, s)\right)-\frac{\partial}{\partial x}(f(x, t) p(x, t | y, s))
$$

## SSA for Fokker-Planck Equation with a reflecting B.C.

1. Generate a normally distributed (with zero mean and unit variance) random number $\xi$.

2. Compute possible position $X(t+\Delta t)$ at time $t+\Delta t$ by eqn.(1)

3. If $X(t+\Delta t)$ computed by (1) is less than $0,$ then set instead
$$
X(t+\Delta t)=-X(t)-f(X(t), t) \Delta t-g(X(t), t) \sqrt{\Delta t} \xi
$$

4. Then continue with step $(\mathrm{a} 7)$ for time $t+\Delta t$

**The Fokker-Planck equation may be thought of as a conservation equation for probability density : it may be written as**

$$
\frac{\partial}{\partial t} p(x, t | y, s)+\frac{\partial Q}{\partial x}=0
$$

where $Q$ is the **probability flux** given by

$$
Q=f(x, t) p(x, t | y, s)-\frac{\partial}{\partial x}\left(\frac{g(x, t)^{2}}{2} p(x, t | y, s)\right)
$$

We see that the reflecting boundary condition corresponds to $Q(0)=0$ as we would expect: **there is no flux of probability out of the domain since trajectories are reflected at the boundary**.

## Kolmogorov Backward Equation

Sometimes we want to know how the likelihood of ending up in a given state depends on the starting state. **In such cases the end position is known, and the starting position is undetermined: in a sense we are thinking about the problem backwards.**

we rename variables in the Chapman–Kolmogorov equation to obtain

$$
p(x, t | y, s-\Delta s)=\int_{\mathbb{R}} p(x, t | z, s) p(z, s | y, s-\Delta s) \mathrm{d} z\text{---(8)}
$$

As before, this equation is valid for any $\Delta s,$ but we will now take the limit in which $\Delta s \rightarrow 0 .$ Using the Taylor expansion about the point $z=y$ we rewrite

$$
\begin{aligned} p(x, t | z, s)=& p(x, t | y, s)+(z-y) \frac{\partial p}{\partial y}(x, t | y, s) \\ &+\frac{(z-y)^{2}}{2} \frac{\partial^{2} p}{\partial y^{2}}(x, t | y, s)+o\left((z-y)^{2}\right) \end{aligned}\text{---(9)}
$$

Substituting into (8) the right-hand side of (9) gives:

$$
\begin{aligned} p(x, t | y, s-\Delta s)=p &(x, t | y, s) \times \int_{\mathbb{R}} p(z, s | y, s-\Delta s) \mathrm{d} z \\ &+\frac{\partial p}{\partial y}(x, t | y, s) \times \int_{\mathbb{R}}(z-y) p(z, s | y, s-\Delta s) \mathrm{d} z \\ &+\frac{\partial^{2} p}{\partial y^{2}}(x, t | y, s) \times \int_{\mathbb{R}} \frac{(z-y)^{2}}{2} p(z, s | y, s-\Delta s) \mathrm{d} z+O\left(\Delta s^{2}\right) \end{aligned}
$$

Using (5),(6), and (7), we have:

$$
\begin{aligned} \frac{p(x, t | y, s-\Delta s)-p(x, t | y, s)}{\Delta s} \\=& f(y, s) \frac{\partial p}{\partial y}(x, t | y, s)+\frac{g(y, s)^{2}}{2} \frac{\partial^{2} p}{\partial y^{2}}(x, t | y, s)+O(\Delta s) \end{aligned}
$$

Passing to the limit $\Delta s \rightarrow 0,$ we derive the so-called **Kolmogorov backward equation**:

$$
-\frac{\partial p}{\partial s}(x, t | y, s)=f(y, s) \frac{\partial p}{\partial y}(x, t | y, s)+\frac{g(y, s)^{2}}{2} \frac{\partial^{2} p}{\partial y^{2}}(x, t | y, s)
$$

To simplify the notation, we now define the diffusion coefficient by:

$$
d(y, s)=\frac{g(y, s)^{2}}{2}
$$

Then the Fokker–Planck equation is

$$
\frac{\partial p}{\partial t}(x, t)=\frac{\partial^{2}}{\partial x^{2}}(d(x, t) p(x, t))-\frac{\partial}{\partial x}(f(x, t) p(x, t))
$$

And the Kolmogorov backward equation

$$
-\frac{\partial p}{\partial s}(x, t | y, s)=f(y, s) \frac{\partial p}{\partial y}(x, t | y, s)+d(y, s) \frac{\partial^{2} p}{\partial y^{2}}(x, t | y, s)
$$

the stationary distribution is given by:

$$
p_{s}(x)=\frac{C}{d(x)} \exp \left[\int_{0}^{x} \frac{f(y)}{d(y)} \mathrm{d} y\right]
$$

where the normalization constant is

$$
C=\left(\int_{\mathbb{R}} \frac{1}{d(x)} \exp \left[\int_{0}^{x} \frac{f(y)}{d(y)} \mathrm{d} y\right] \mathrm{d} x\right)^{-1}
$$
