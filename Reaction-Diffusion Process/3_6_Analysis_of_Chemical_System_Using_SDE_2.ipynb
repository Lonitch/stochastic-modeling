## Chemical Fokker-Planck Equation
Let's come back to the following system.

$$
2 A+B \stackrel{k_{1}}{\longrightarrow} 3 A
$$
$$
\emptyset \overset{k2}{\underset{k_{3}}\leftrightarrow} A
$$
$$
\emptyset \stackrel{k_{4}}{\longrightarrow} B
$$

Let $p(x, y, t)$ be the probability that $A(t)=x$ and $B(t)=y .$ As before, we smoothly extend the values of $p(x, y, t)$ to non-integer values of $x$ and $y .$ The propensity functions are:

$$
\begin{aligned} \alpha_{1}(x, y) &=x(x-1) y k_{1} / v^{2}, & & \alpha_{2}(x, y)=k_{2} v \\ \alpha_{3}(x, y) &=x k_{3}, & & \alpha_{4}(x, y) &=k_{4} v \end{aligned}
$$

Again, use the generalized chemical Fokker-Planck equation, we have:

$$
\begin{aligned} \frac{\partial}{\partial t} p(x, y, t)=& \frac{\partial}{\partial x}\left[\left(-x(x-1) y k_{1} / v^{2}-k_{2} v+x k_{3}\right) p(x, y, t)\right] \\ &+\frac{\partial}{\partial y}\left[\left(x(x-1) y k_{1} / v^{2}-k_{4} v\right) p(x, y, t)\right] \\ &+\frac{1}{2} \frac{\partial^{2}}{\partial x^{2}}\left[\left(x(x-1) y k_{1} / v^{2}+k_{2} v+x k_{3}\right) p(x, y, t)\right] \end{aligned}
$$
$$
\begin{array}{l}{+\frac{1}{2} \frac{\partial^{2}}{\partial y^{2}}\left[\left(x(x-1) y k_{1} / v^{2}+k_{4} v\right) p(x, y, t)\right]} \\ {-\frac{\partial^{2}}{\partial x \partial y}\left[x(x-1) y k_{1} / v^{2} p(x, y, t)\right]}\end{array}
$$

## Constant-B Approximation

This equation is complicated to analyse, so we adopt a more intuitive approach here. We know from notebook(2.2) that the number of B molecules is slowly growing while the number of A fluctuates rapidly. Let's suppose that **the period of the oscillation is controlled by the fluctuations in A.**

We first **fix the number of B=b.** The system now becomes

$$
2 A \stackrel{k_{1}b/v}{\longrightarrow} 3 A
$$
$$
\emptyset \overset{k2}{\underset{k_{3}}\leftrightarrow} A
$$

And the corresponding Fokker-Planck equation gives:

$$
\begin{array}{l}{f(x ; b)=\left(k_{1} / v^{2}\right) x(x-1) b+k_{2} v-k_{3} x} \\ {d(x ; b)=\left[\left(k_{1} / v^{2}\right) x(x-1) b+k_{2} v+k_{3} x\right] / 2}\end{array}\text{---(1)}
$$

Let $\tau(x ; b)$ be the average time for $A(t)$ to leave the interval $\left[0,10^{3}\right)$ provided that initially $A(0)=x .$ Using equation(8) in the notebook(3.3), the mean exit time $\tau(x ; b)$ can be computed by

$$
\tau(x ; b)=\int_{x}^{1000} \frac{1}{d(z ; b)} \int_{0}^{z} \exp [\Phi(z ; b)-\Phi(y ; b)] \mathrm{d} y \mathrm{d} z
$$

where

$$
\Phi(x ; b)=-\int_{0}^{x} \frac{f(y ; b)}{d(y ; b)} \mathrm{d} y+\log (d(x ; b))\text{---(2)}
$$

The local extrema of the potential $\Phi(x ; b)$ can be found as the solutions of $\Phi^{\prime}(x ; b)=0$. Differentiating (2), we obtain:

$$
-f(x ; b)+d^{\prime}(x ; b)=0
$$

and its two solutions:

$$
\begin{array}{l}{x_{f}(b)=1+\frac{k_{3} v^{2}}{2 k_{1} b}-\frac{1}{2} \sqrt{\left(2+\frac{k_{3} v^{2}}{k_{1} b}\right)^{2}-\frac{2 k_{1} b+4 k_{2} v^{3}-2 k_{3} v^{2}}{k_{1} b}}} \\ {x_{u}(b)=1+\frac{k_{3} v^{2}}{2 k_{1} b}+\frac{1}{2} \sqrt{\left(2+\frac{k_{3} v^{2}}{k_{1} b}\right)^{2}-\frac{2 k_{1} b+4 k_{2} v^{3}-2 k_{3} v^{2}}{k_{1} b}}}\end{array}
$$

**Roughly speaking, the values of $A$ in the interval $\left(0, x_{u}(b)\right)$ correspond to fluctuations around the locally favourable state $x_{f}(b)$ and the values of $A$ larger than $x_{u}(b)$ correspond to the excited state.** This stationary distribution can then be written as:

$$
p_{s}(x ; b)=\exp [-\Phi(x ; b)]\left(\int_{0}^{x_{u}(b)} \exp [-\Phi(z ; b)] \mathrm{d} z\right)^{-1}
$$

and

$$
\tau(b)=\int_{0}^{x_{u}(b)} p_{s}(x ; b) \tau(x ; b) \mathrm{d} x
$$

## Slowly-Varying B Approximation

We now allow B to vary slowly. **Since the fluctuations in B are much smaller than those in A, we model the
change in b deterministically by using the following ODE:**

$$
\frac{\mathrm{d} b}{\mathrm{d} t}=w(b)\text{---(3)}
$$
$$
w(b)=-\frac{k_{1} b}{v^{2}} \int_{0}^{x_{u}(b)} x(x-1) p_{s}(x ; b) \mathrm{d} x+k_{4} v
$$
In particular, the average time to reach $B(t)=b$ provided that $B(0)=0$ can be computed (under the condition that the system did not fire) as
$$
\int_{0}^{b} \frac{1}{w(z)} \mathrm{d} z
$$

For a given $b$, the probability to leave the $A$
interval $\left(0,10^{3}\right)$ (i.e. the probability to fire) during the time interval $[t, t+d t)$ is equal to dt $\tau(b) .$ So we conclude that the probability to leave the $A$ interval $\left(0,10^{3}\right)$ during the infinitesimally small $B$ interval $[b, b+d b)$ is
equal to
$$
\frac{\mathrm{d} b}{\tau(b) w(b)}
$$

$$
\begin{array}{l}{\text { and the probability to arrive at } b \text { without previously firing is therefore equal to }} \\ {\qquad \exp \left[-\int_{0}^{b} \frac{\mathrm{d} z}{\tau(z) w(z)}\right]}\end{array}
$$

Now, we can estimate the period of oscillations as

$$
\tau_{p e r} \approx \int_{1000}^{10000} \frac{1}{\tau(b) w(b)}\left(\int_{0}^{b} \frac{1}{w(z)} \mathrm{d} z\right) \exp \left[-\int_{0}^{b} \frac{\mathrm{d} z}{\tau(z) w(z)}\right] \mathrm{d} b
$$
