## Overview

Let's come back to the following SDE:

$$
X(t+\mathrm{d} t)=X(t)+\left(-k_{1} X(t)^{3}+k_{2} X(t)^{2}-k_{3} X(t)+k_{4}\right) \mathrm{d} t+k_{5} \mathrm{d} W \text{---(1)}
$$

Its deterministic description is:

$$
\frac{\mathrm{d} x}{\mathrm{d} t}=-k_{1} x^{3}+k_{2} x^{2}-k_{3} x+k_{4}\text{---(2)}
$$

where we obtain two stable states $x_{s1}=100$,$x_{s2}=400$ and a unstable state $x_{u}=250$. We denote by $\bar{\tau}$ **the average time for simulation to reach the point $x_{u}$ provided that it started at $X(0)=x_{s 1}$.** If the simulation reaches $x_{u},$ there is a $50 \%$ chance that it will return back to $x_{s 1}$ and a $50 \%$ chance to continue on to $x_{s 2}$. **Thus the mean switching time from $x_{s 1}$ to $x_{s 2}$ is simply $2 \bar{\tau} .$**

We can estimate $\bar{\tau}$ using SSA as follows:

- We start at $X(0)=x_{s 1}$

- We wait until the first time the simulation leaves the interval $\left(-\infty, x_{u}\right)$.

- We calculate the mean value of $\bar{\tau}$ over multiple realizations.

## Analytical formula for $\bar{\tau}$

We first let $f(x, t) \equiv f(x)$ and $g(x, t) \equiv g(x)$ in the following eqn:

$$
X(t+\mathrm{d} t)=X(t)+f(X(t), t) \mathrm{d} t+g(X(t), t) \mathrm{d} W\text{---(3)}
$$

**Let $h(y, t)$ be the probability that $X\left(t^{\prime}\right) \in\left(-\infty, x_{u}\right)$ for all times $0<t^{\prime}<t$ given that it started at
$X(0)=y \in\left(-\infty, x_{u}\right) .$** Then

$$
h(y, t)=\int_{-\infty}^{x_{u}} p(x, t | y, 0) \mathrm{d} x\text{---(4)}
$$

Distribution $p$ satisfies the Fokker-Planck and Kolmogorov backward equations with the boundary conditions $p\left(x_{u}, t | y, s\right)=$
$p\left(x, t | x_{u}, s\right)=0,$ so that $p(x, t | y, s)=0$ if $y \geq x_{u}$ or $x \geq x_{u} .$ since we assume that the coefficients $f(x)$ and $g(x)$ of the $\operatorname{SDE}(3)$ do not depend explicitly on
time, **we can shift time in $p$ in (4) to give**

$$
h(y, t)=\int_{-\infty}^{x_{u}} p(x, 0 | y,-t) \mathrm{d} x
$$

Using the transformation $s$ to $-t$ in the Kolmogorov backward equation
$$
\frac{\partial p}{\partial t}(x, 0 | y,-t)=f(y) \frac{\partial p}{\partial y}(x, 0 | y,-t)+d(y) \frac{\partial^{2} p}{\partial y^{2}}(x, 0 | y,-t)
$$
Integrating over x

$$
\frac{\partial h}{\partial t}(y, t)=f(y) \frac{\partial h}{\partial y}(y, t)+d(y) \frac{\partial^{2} h}{\partial y^{2}}(y, t) \text{---(5)}
$$

**The probability that $X$ first leaves the interval $\left(-\infty, x_{u}\right)$ during the time interval $[t, t+d t)$ (under the condition that $X(0)=y$ ) is given by**

$$
h(y, t)-h(y, t+\mathrm{d} t) \approx-\frac{\partial h}{\partial t}(y, t) \mathrm{d} t
$$

Thus,

$$
\tau(y)=-\int_{0}^{\infty} t \frac{\partial h}{\partial t}(y, t) \mathrm{d} t=\int_{0}^{\infty} h(y, t) \mathrm{d} t\text{---(6)}
$$

Now, we use (6) to integrate (5) over $t$. Using the fact that $h(y,0)=1$ and $h(y,\infty)=0$, we have:

$$
-1=f(y) \frac{\mathrm{d} \tau}{\mathrm{d} y}(y)+d(y) \frac{\mathrm{d}^{2} \tau}{\mathrm{d} y^{2}}(y) \quad \text { for } y \in\left(-\infty, x_{u}\right)\text{---(7)}
$$

With the following B.C.

$$
\frac{\mathrm{d} \tau}{\mathrm{d} y}(-\infty)=0
$$
and
$$
\tau\left(x_{u}\right)=0
$$

We solve (7) giving:

$$
\tau(y)=\int_{y}^{x_{u}} \frac{1}{d(z) p_{s}(z)} \int_{-\infty}^{z} p_{s}(x) \mathrm{d} x \mathrm{d} z\text{---(8)}
$$
Thus $\bar{\tau}$ can be computed as

$$
\bar{\tau}=\tau\left(x_{s 1}\right)=\int_{x_{s 1}}^{x_{u}} \frac{1}{d(z) p_{s}(z)} \int_{-\infty}^{z} p_{s}(x) \mathrm{d} x \mathrm{d} z
$$

Since we already know $p_s$ in the note (3.1), we now have the analytical value for $\bar{\tau}=59.45$.
