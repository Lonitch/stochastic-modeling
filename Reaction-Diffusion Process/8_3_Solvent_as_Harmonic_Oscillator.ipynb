## Overview
---
In this notebook, we assume that the diffusing particle of interest is coupled with $N$ solvent molecules. Then the position $X(t)$ and velocity $V(t)$ of the diffusing particle evolve according to:

$$
\begin{aligned} \frac{\mathrm{d} X}{\mathrm{d} t} &=V(t)\text{---(1)} \\ m_{p} \frac{\mathrm{d} V}{\mathrm{d} t}(t) &=\sum_{i=1}^{N} k_{i}\left(x_{i}(t)-X(t)\right)\text{---(2)} \end{aligned}
$$

We assume that the $i$ th solvent molecule has mass $m_{s, i}$ Using Newton's second law of motion, we get

$$
\begin{aligned} \frac{\mathrm{d} x_{i}}{\mathrm{d} t}(t) &=v_{i}(t)\text{---(3)} \\ m_{s, i} \frac{\mathrm{d} v_{i}}{\mathrm{d} t}(t) &=k_{i}\left(X(t)-x_{i}(t)\right)\text{---(4)} \end{aligned}
$$

In the (4), we are no longer including the random and frictional forces, because we **explicitly model all solvent molecules.**

For each $i,$ the solvent equations $(3)-(4)$ describe a forced harmonic oscillator: each of the solvent molecules has a natural frequency of oscillation $\omega_{i}=\sqrt{k_{i} / m_{s, i}} $ These equations can be solved to get:

$$
x_{i}(t)=x_{i}(0) \cos \left(\omega_{i} t\right)+\frac{v_{i}(0)}{\omega_{i}} \sin \left(\omega_{i} t\right)+\omega_{i} \int_{0}^{t} \sin \left(\omega_{i}(t-\tau)\right) X(\tau) \mathrm{d} \tau\text{---(5)}
$$

Using integration by parts and equation (1), we have:

$$
\omega_{i} \int_{0}^{t} \sin \left(\omega_{i}(t-\tau)\right) X(\tau) \mathrm{d} \tau=X(t)-\cos \left(\omega_{i} t\right) X(0)-\int_{0}^{t} \cos \left(\omega_{i}(t-\tau)\right) V(\tau) \mathrm{d} \tau
$$

Substituting into (5) and assuming without loss of generality that the diffusing particle started at the origin, we obtain:

$$
x_{i}(t)-X(t)=x_{i}(0) \cos \left(\omega_{i} t\right)+\frac{v_{i}(0)}{\omega_{i}} \sin \left(\omega_{i} t\right)-\int_{0}^{t} \cos \left(\omega_{i}(t-\tau)\right) V(\tau) \mathrm{d} \tau
$$

Substituting into (2), we have:

$$
\frac{\mathrm{d} V}{\mathrm{d} t}(t)=\sum_{i=1}^{N} \frac{k_{i}}{m_{p}}\left(x_{i}(0) \cos \left(\omega_{i} t\right)+\frac{v_{i}(0)}{\omega_{i}} \sin \left(\omega_{i} t\right)-\int_{0}^{t} \cos \left(\omega_{i} \tau\right) V(t-\tau) \mathrm{d} \tau\right)
$$

Thus the evolution of the diffusing particle can again be formally written in the form of the generalized Langevin equation provided that the friction kernel $\kappa(\tau)$ and noise term $R(t)$ are given by:

$$
\begin{aligned} \kappa(\tau) &=\frac{1}{m_{p}} \sum_{i=1}^{N} m_{s, i} \omega_{i}^{2} \cos \left(\omega_{i} \tau\right) \text{---(6)}\\ R(t) &=\frac{1}{m_{p}} \sum_{i=1}^{N} x_{i}(0) m_{s, i} \omega_{i}^{2} \cos \left(\omega_{i} t\right)+v_{i}(0) m_{s, i} \omega_{i} \sin \left(\omega_{i} t\right) \text{---(7)}\end{aligned}
$$

where masses $m_{s, i}$ and frequencies $\omega_{i}=\sqrt{k_{i} / m_{s, i}}$ are yet to be specified.

## Choice of Initial Condition

The dynamics of the diffusing particle will also depend on our choice of the initial positions and velocities of solvent oscillators. Since we do not have much information about them, it is natural to assume that $x_i(0)$ and $v_i(0)$ are both independently sampled according to some probability distributions. **If we sample them according to their equilibrium distributions, then the noise autocorrelation function can be expressed as**

$$
\langle R(t) R(t-\tau)\rangle=\langle R(\tau) R(0)\rangle=\frac{1}{m_{p}^{2}} \sum_{i=1}^{N}\left\langle x_{i}^{2}(0)\right\rangle m_{s, i}^{2} \omega_{i}^{4} \cos \left(\omega_{i} \tau\right)\text{---(8)}
$$

where we have used the fact that $x_{i}(0)$ and $v_{i}(0)$ are independent and have zero
mean, i.e. $\left\langle x_{i}(0)\right\rangle=\left\langle v_{i}(0)\right\rangle= 0 .$ If we associate the kinetic energy of a particle with the absolute temperature T through the relation $m_{s, i}\left\langle v_{i}^{2}(0)\right\rangle / 2=k_{B} T / 2 .$ The average potential energy of a harmonic oscillator, $k_{i}\left\langle x_{i}^{2}(0)\right\rangle / 2,$ is also equal to the same value $k_{B} T / 2 .$ Thus, we have

$$
\left\langle x_{i}^{2}(0)\right\rangle=\frac{k_{B} T}{m_{s, i} \omega_{i}^{2}} \quad \text { and } \quad\left\langle v_{i}^{2}(0)\right\rangle=\frac{k_{B} T}{m_{s, i}}
$$

Substituting (8) into (6), we obtain the **generalized fluctuation-dissipation theorem** in the form:

$$
\langle R(t) R(t-\tau)\rangle=\frac{k_{B} T}{m_{p}^{2}} \sum_{i=1}^{N} m_{s, i} \omega_{i}^{2} \cos \left(\omega_{i} \tau\right)=\frac{k_{B} T}{m_{p}} \kappa(\tau)
$$

## Use SSA from Notebook(8.2)

To use the properties of the generalized Langevin equation, **we have to choose the masses $m_{s,i}$ and frequencies $\omega_i=\sqrt{k_i/m_{s,i}}$ in such a way that the friction kernel (6) satisfies $\kappa(\tau) \rightarrow 0$ as $\tau \rightarrow \infty$.

If we assume that all spring constants $k_i,i=1,2,...,N,$ are equal to the same value $k$, i.e.,
$$
k_{i}=m_{s, i} \omega_{i}^{2}=k, \quad \text { where } \quad k=\frac{2 \beta m_{p} \omega_{\mathrm{m}}}{N \pi}\text{---(9)}
$$
Here, $\beta$ is a new parameter. Using (9), the friction kernel simplifies to:

$$
\kappa(\tau)=\frac{2 \beta \omega_{\mathrm{m}}}{\pi} \frac{1}{N} \sum_{i=1}^{N} \cos \left(\omega_{i} \tau\right)
$$
Letting $N \rightarrow \infty,$ we can replace this sum by an integral,
$$
\kappa(\tau)=\frac{2 \beta \omega_{\mathrm{m}}}{\pi} \int_{0}^{\infty} \cos (\omega \tau) p(\omega) \mathrm{d} \omega\text{---(10)}
$$

**where $p(\omega)$ is the probability distribution from which we sample the frequencies $\omega_{i} .$** A simple choice is to sample $\omega_{i}$ uniformly in interval $\left[0,2 \omega_{\mathrm{m}}\right] .$ Then $\omega_{\mathrm{m}}$ is indeed the mean frequency and equation (10) implies

$$
\kappa_{\mathrm{u}}(\tau)=\frac{\beta}{\pi} \int_{0}^{2 \omega_{\mathrm{m}}} \cos (\omega \tau) \mathrm{d} \omega=\frac{\beta \sin \left(2 \omega_{\mathrm{m}} \tau\right)}{\pi \tau}
$$

We observe that $\kappa_{\mathrm{u}}(0) \rightarrow \infty$ as $\omega_{\mathrm{m}} \rightarrow \infty,$ and we also have
$$
\int_{0}^{\infty} \kappa_{\mathrm{u}}(\tau) \mathrm{d} \tau=\frac{\beta}{2}
$$

We can also **sample $\omega_{i}$ from exponential
distribution with mean $\omega_{\mathrm{m}}$. Then equation (10) implies**

$$
\begin{array}{l}{\kappa_{\mathrm{e}}(\tau)=\frac{2 \beta}{\pi} \int_{0}^{\infty} \cos (\omega \tau) \exp \left(-\frac{\omega}{\omega_{\mathrm{m}}}\right) \mathrm{d} \omega=\frac{2 \beta}{\pi} \frac{\omega_{\mathrm{m}}}{\omega_{\mathrm{m}}^{2} \tau^{2}+1}} \\ {\int_{0}^{\infty} \kappa_{\mathrm{e}}(\tau) \mathrm{d} \tau=\beta}\end{array}
$$
