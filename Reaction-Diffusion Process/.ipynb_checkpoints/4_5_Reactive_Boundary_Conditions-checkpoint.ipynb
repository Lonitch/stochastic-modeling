## Overview

A partially adsorbing boundary condition for SSAs can be formulated as follows: **whenever a molecule hits the boundary, it is adsorbed with some probability, and otherwise it is reflected**

The key question is: **What is the probability that the particle is adsorbed rather than reflected, and how does this relate to the reactive properties of the boundary for a given stochastic model?**

To answer this question, let's consider a chemical species A diffusing in the half space $\{x>0\}=(0, \infty) \times \mathbb{R} \times \mathbb{R}$ We suppose that the boundary $\{x=0\}$ is uniformly covered by receptors $B .$ Molecules of $A$ and $B$ react according to the second-order chemical reaction:

$$
A+B \stackrel{k}{\longrightarrow} \emptyset
$$

Since the receptors are uniformly distributed on the boundary, there will be no concentration differences in the y- and z-directions. Thus we focus on the behaviour along the x-axis only.

Assuming that we have a lot of molecules of A in the system, we can describe the system in terms of the density, so that so that $\varrho(x, t) \delta x$ gives the number of molecules in the small box $[x, x+\delta x) \times[0,1] \times[0,1]$ at time $t$. The time evolution of $\varrho(x, t)$ is governed by the diffusion equation

$$
\frac{\partial \varrho}{\partial t}=D \frac{\partial^{2} \varrho}{\partial x^{2}}
$$

Let us now consider a small box $[0, \delta x)\times[0,1] \times[0,1]$. Let $b(t)$ be the number of available resceptors B per unit of area of the boundary $\{x=0\}$. Thus, **the probability that the chemical reaction takes place in the time interval **$[t,t+dt)$ is equal to 

$$
(\varrho(0, t) \delta x b(t) k / \delta x) \mathrm{d} t=\sigma(t) \varrho(0, t) \mathrm{d} t\text{---(1)}
$$

where $\sigma=kb(t)$. We will assume that $\sigma$ is independent of time (*the number of receptors on the boundary is in excess of the number of diffusing molecules*). The **diffusive flux** through the boundary (i.e. the average number of molecules that are removed by the boundary per unit area and per unit time) is thus **balanced by the rate of reaction**:

$$
D \frac{\partial \varrho}{\partial x}(0, t)=\sigma \varrho(0, t)\text{---(2)}
$$

which is the so-called **Robin boundary condition** (or reactive boundary condition) at x = 0. The non-negative constant $\sigma$ (which has units $\mathrm{mm} \mathrm{sec}^{-1}$ ) describes the reactivity of the boundary. A reflective boundary corresponds to $\sigma=0$ and a fully adsorbing boundary corresponds to $\sigma=\infty .$

## Connection btw Robin B.C. and Stochastic Simulation

For definiteness we consider the behaviour of molecules in a one-dimensional interval $[0,L]$ with a Robin boundary condition at $x = 0$ and a reflective boundary condition at $x = L$.

To implement the reactive boundary condition at $x = 0$, we simply add another reaction (a jump from the first compartment to the left where the molecules are removed from the solution):

$$
A_{1} \stackrel{\sigma / h}{\rightarrow} \emptyset
$$

To see this, we first consider a compartment simulation based on Robin B.C. Let us define the operators $R_{i}, L_{i}: \mathbb{N}^{K} \rightarrow \mathbb{N}^{K}$ (where $\mathbb{N}$ is the set of non-negative integers) by:

$$
R_{i}:\left[n_{1}, \dots, n_{i}, n_{i+1}, \dots, n_{K}\right] \rightarrow\left[n_{1}, \dots, n_{i}+1, n_{i+1}-1, \dots, n_{K}\right]
$$
for $i=1,2, \ldots, K-1,$ and
$$
L_{i}:\left[n_{1}, \ldots, n_{i-1}, n_{i}, \ldots, n_{K}\right] \rightarrow\left[n_{1}, \ldots, n_{i-1}-1, n_{i}+1, \ldots, n_{K}\right]
$$

The master equation which corresponds to the diffusion process can be given by:

$$
\begin{aligned} \frac{\partial p(\mathbf{n})}{\partial t}=& d \sum_{j=1}^{K-1}\left\{\left(n_{j}+1\right) p\left(R_{j} \mathbf{n}\right)-n_{j} p(\mathbf{n})\right\} \\ &+d \sum_{j=2}^{K}\left\{\left(n_{j}+1\right) p\left(L_{j} \mathbf{n}\right)-n_{j} p(\mathbf{n})\right\} \\ &+\frac{\sigma}{h}\left\{\left(n_{1}+1\right) p\left(n_{1}+1, n_{2}, n_{3}, \ldots, n_{K}\right)-n_{1} p(\mathbf{n})\right\} \end{aligned}\text{---(3)}
$$

The mean is then defined as:

$$
M_{i}(t)=\sum_{\mathbf{n}} n_{i} p(\mathbf{n}, t) \equiv \sum_{n_{1}=0}^{\infty} \sum_{n_{2}=0}^{\infty} \cdots \sum_{n_{K}=0}^{\infty} n_{i} p(\mathbf{n}, t)\text{---(4)}
$$

gives the mean number of molecules in the $i$ th compartment, $i=1,2, \ldots, K$. For the compartments that are far away from the boundary, we multiplying (3) by $n_i$ and summing over $\mathcal{n}$, we obtain a system of equations for $M_i$ of the form:

$$
\frac{\partial M_{i}}{\partial t}=d\left(M_{i+1}+M_{i-1}-2 M_{i}\right), \quad i=2,3, \ldots, K-1\text{---(5)}
$$

for last compartment, we have
$$
\frac{\partial M_{K}}{\partial t}=d\left(M_{K-1}-M_{K}\right)\text{---(6)}
$$

**To determine the boundary value $M_1$ we multiply (3) by $n_1$ and sum over $\mathcal{n}$ to give:**
$$
\frac{\partial M_{1}}{\partial t}=d\left(M_{2}-M_{1}\right)-\frac{\sigma M_{1}}{h}\text{---(7)}
$$

**If we approximate $M_{k}(t) / h$ with $\varrho\left(x_{k}, t\right)$**, and identifying the mid-point of kth compartment is $x_{k}=k h+h / 2$, we can transform the (7) into the following form:

$$
h \frac{\partial \varrho}{\partial t}(h / 2, t)=D \frac{\varrho(3 h / 2, t)-\varrho(h / 2, t)}{h}-\sigma \varrho(h / 2, t)
$$
Taylor expanding $\varrho(3 h / 2, t)$ about the point $x=h / 2$ gives
$$
\varrho(3 h / 2, t)=\varrho(h / 2, t)+h \frac{\partial \varrho}{\partial x}(h / 2, t)+O\left(h^{2}\right)
$$

so we finally arrive at the Robin B.C.
$$
D \frac{\partial \varrho}{\partial x}(h / 2, t)=\sigma \varrho(h / 2, t)+O(h)
$$

We have seen that molecules are removed from the first compartment with the rate $\sigma / h,$ while the diffusion jumps to the left occur with the rate $d=D / h^{2}$. Consequently, **the probability of the removal of a molecule that has hit the boundary $x=0$ is a conditional probability and is equal to**

$$
P_{1}(h)=\frac{\sigma / h}{d}=\frac{\sigma h}{D}
$$

where $h$ has to be chosen sufficiently small to ensure that $P_{1}(h)$ is less than 1. Thus the boundary reaction can be reformulated as the following partially adsorbing boundary condition at $x=0:$ **whenever a molecule hits the boundary, it is adsorbed with the probability $P_{1}(h),$ and otherwise it is rejected**.

## Diffusion Modeled by SDE+Partially Adsorbing Boundary

The partially adsorbing boundary condition is implemented using the following two steps:

1. If $X(t+\Delta t)$ computed is negative then $X(t+\Delta t)=-X(t)-\sqrt{2 D \Delta t} \xi$ with probability $1-P_{2}(\Delta t)$, otherwise we remove the molecule from the system.

2. If $X(t+\Delta t)$ computed is positive then we remove the molecule from the system with probability $exp[-X(t)X(t+\Delta t)/(D\Delta t)]P_{2} ( \Delta t )$

The steps above leads to the Robin boundary condition if we choose (Erban and Chapman, 2007):
$$
P_{2}(\Delta t)=\frac{\sigma \sqrt{\pi \Delta t}}{2 \sqrt{D}}
$$

### Velocity-Jump process
The Robin boundary condition can also be derived for a variant of the velocity-jump process:

$$
\begin{array}{l}{X(t+\Delta t)=X(t)+V(t) \Delta t} \\ {V(t+\Delta t)=V(t)-\beta V(t) \Delta t+\beta \sqrt{2 D \Delta t} \xi}\end{array}
$$

which models a reactive boundary in terms of a boundary probability 
$$
P_{4}(\beta)=\frac{\sigma \sqrt{2 \pi}}{\sqrt{D \beta}}
$$

The process is reformulated as: whenever a molecule hits the boundary, it is adsorbed with probability $P_{4}(\beta),$ and otherwise it is reflected. The Robin boundary condition is recovered in the limit $\beta \rightarrow \infty$.

```python
## Implement Vel-jump + Robin boundary at the left
import numpy as np
from numpy.random import uniform
from scipy.stats import norm

def jump_vel1(beta,D,sigma,dt,L,ttot,x0,v0):
    v = v0
    xlst = [x0]
    tlst = [0]
    rv = norm()
    p4 = sigma*np.sqrt(2*np.pi)/np.sqrt(D*beta)
    while tlst[-1]<ttot:
        tlst.append(tlst[-1]+dt)
        cur = xlst[-1]+v*dt
        
        if cur<0:
            xlst.append(cur)
            return tlst,xlst
        
        elif cur>L:
            xlst.append(2*L-cur)
            
        else:
            xlst.append(cur)
            p = np.exp(-xlst[-2]*xlst[-1]/(D*dt))*p4
            if uniform()<p:
                return tlst,xlst
            
        v = v-beta*v*dt+beta*np.sqrt(2*D*dt)*rv.rvs()
    return tlst,xlst
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
param=[21,0.0001,0.001,0.1,1,540,0.5,0.001]
rep = 10
for i in range(10):
    x,y = jump_vel1(*param)
    plt.plot(x,y)
plt.xlabel('time[s]')
plt.ylabel('x[mm]')
```
