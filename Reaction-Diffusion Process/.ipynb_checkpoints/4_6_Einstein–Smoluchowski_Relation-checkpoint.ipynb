In this notebook, we derive Einstein-Smoluchowski relation by using the following two SDE:

$$
\begin{array}{l}{X(t+\Delta t)=X(t)+V(t) \Delta t}\text{---(1)} \\ {V(t+\Delta t)=V(t)-\beta V(t) \Delta t+\beta \sqrt{2 D \Delta t} \xi}\text{---(2)}\end{array}
$$

We first multiply (2) by $X(t)$ to give:
$$
V(t+\Delta t) X(t)-V(t) X(t)=-\beta V(t) X(t) \Delta t+\beta \sqrt{2 D \Delta t} X(t) \xi
$$

Using (1), we can rewrite this as:
$$
\begin{array}{l}{V(t+\Delta t) X(t+\Delta t)-V(t) X(t)-V(t+\Delta t) V(t) \Delta t} \\ {\quad=-\beta V(t) X(t) \Delta t+\beta \sqrt{2 D \Delta t} X(t) \xi}\end{array}
$$

Using the following definition:
$$
V(t)=\frac{\mathrm{d} X}{\mathrm{d} t}(t)
$$
and dividing by $\Delta t$ gives:

$$
\frac{1}{2 \Delta t}\left(\frac{\mathrm{d} X^{2}}{\mathrm{d} t}(t+\Delta t)-\frac{\mathrm{d} X^{2}}{\mathrm{d} t}(t)\right)-V(t+\Delta t) V(t)=-\frac{\beta}{2} \frac{\mathrm{d} X^{2}}{\mathrm{d} t}(t)+\beta \sqrt{\frac{2 D}{\Delta t}} X(t) \xi
$$

Now we take an average to give:

$$
\frac{1}{\Delta t}\left(\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t+\Delta t)-\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)\right)-2\langle V(t+\Delta t) V(t)\rangle=-\beta \frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)
$$

Passing to the limit $\Delta t \rightarrow 0$ gives

$$
\frac{1}{\beta} \frac{\mathrm{d}^{2}\left\langle X^{2}\right\rangle}{\mathrm{d} t^{2}}(t)+\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)=\frac{2\left\langle V(t)^{2}\right\rangle}{\beta}\text{---(3)}
$$

The kinetic energy associated with the movement of a small particle along the x-axis is equal to $k_BT/2$. Thus $m_{p}\left\langle V(t)^{2}\right\rangle / 2=k_{B} T / 2$, where $m_{p}$ is the particle mass. Consequently, (3) becomes:

$$
\frac{1}{\beta} \frac{\mathrm{d}^{2}\left\langle X^{2}\right\rangle}{\mathrm{d} t^{2}}(t)+\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)=\frac{2 k_{B} T}{\beta m_{p}}
$$

**since $\beta$ is a large parameter, we can (for times  drop the second-order time derivative to give our first expression for the evolution of $\left\langle X^{2}\right\rangle$**

$$
\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)=\frac{2 k_{B} T}{\beta m_{p}}\text{---(4)}
$$

To get an alternative expression, we use the diffusion SDE:

$$
X(t+\Delta t)=X(t)+\sqrt{2 D \Delta t} \xi
$$

If we take the square of the equation above, we get:

$$
X(t+\Delta t)^{2}=X(t)^{2}+2 X(t) \sqrt{2 D \Delta t} \xi+2 D \Delta t \xi^{2}
$$

Taking averages (again using the fact that $\xi$ is normally distributed with zero mean and unit variance), and dividing by $\Delta t,$ we get:

$$
\frac{\left\langle X(t+\Delta t)^{2}\right\rangle-\left\langle X(t)^{2}\right\rangle}{\Delta t}=2 D
$$

Passing to the limit $\Delta t \rightarrow 0$ gives our second expression for the evolution of $\left\langle X^{2}\right\rangle$:

$$
\frac{\mathrm{d}\left\langle X^{2}\right\rangle}{\mathrm{d} t}(t)=2 D
$$

Compare (4) and (5), we get Einsteinâ€“Smoluchowski relation:
$$
D=\frac{k_{B} T}{\gamma}\text{---(6)}
$$

where $\gamma=\beta m_{p}$. The coefficient $\gamma$ is called the frictional drag coefficient, and **it is, unlike $\beta$, independent of the particle mass $m_p$.** To see this, we multiply (2) by $m_p$ to obtain:

$$
m_{p}(V(t+\mathrm{d} t)-V(t))=-\gamma V(t) \mathrm{d} t+\gamma \sqrt{2 D} \mathrm{d} W
$$

If we formally divide this equation by dt and pass to the limit d $\rightarrow 0$, we get

$$
m_{p} \frac{\mathrm{d} V}{\mathrm{d} t}(t)=-\gamma V(t)+\gamma \sqrt{2 D}\{\frac{dW}{dt}\}
$$
There are two forces on the right-hand side: the random
force and frictional force $F_{f}(t)=-\gamma V(t) ;$ we see that $\gamma$ is indeed the coefficient that relates the frictional drag force $F_{f}(t)$ to the particle velocity $V(t)$. The derivative in the bracket highlights that this is not a derivative in usual sense.

**The value of $\gamma$ depends on the geometry of the object and properties of the surrounding fluid.** If the diffusing particle is a sphere of radius $r$ diffusing in
incompressible viscous liquid, **Stokes law** says that

$$
\gamma=6 \pi \eta r
$$

where $\eta$ is the coefficient of viscosity. Substituting into $(6),$ we obtain the **Einstein (or Stokes-Einstein) relation**:

$$
D=\frac{k_{B} T}{6 \pi \eta r}
$$
