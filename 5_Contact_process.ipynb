## Overview

In this note, we are going to prove that the contact process **exhibits a phase transition:** there is a **critical birth parameter** above which the process starting with one individual survives with positive probability but below which it goes extinct eventually with probability one.

There are three important properties of the contact process:

- **Monotonicity**— The probability of survival starting from a given configuration is nondecreasing with respect to the birth parameter. As for percolation models, this implies *the existence of at most one phase transition at some critical value.*

- **Attractiveness**--The space-time set of occupied sites is also nondecreasing for the inclusion with respect to the initial configuration. A consequence of this property known as attractiveness is that the process starting from the all occupied configuration *converges in distribution to an invariant measure* called the **upper invariant** measure since this is the largest one.

- **Self-duality**—Keeping track of the potential ancestors of a given space-time point going backward in time through the graphical representation defines a new process called the **dual process**. *This process turns out to have the same distribution as the contact process itself*.

Due to the inclusion of local interactions, **the critical value is strictly larger than one**,which contrasts with nonspatial and spatially implicit invasion models. We identify the configuration of the process at time $t$ with the **set of occupied sites** at that time:

1. The process **dies out** when $\xi_{t}=\varnothing$ for some $t$

2. The process survives weakly when $\xi_{t} \neq \varnothing$ for all $t$

3. The process survives strongly when $0 \in \xi_{t}$ infinitely often.

## Monotonicity and attractiveness

### Lemma of Monotonicity
For all $(x, t) \in \mathbb{Z}^{d} \times \mathbb{R}_{+}$

$$
P\left(x \in \xi_{t}^{1}\right) \leq P\left(x \in \xi_{t}^{2}\right) \quad \text { when } \quad \xi_{0}^{1}=\xi_{0}^{2} \quad \text { and } \quad \beta_{1} \leq \beta_{2}
$$

where $\xi_t^1$ and $\xi_t^2$ are two realizations of the same system from the same initial configuration.

It directly follows from the lemma that the probability of weak/strong survival of the process starting from a single individual is nondecreasing with respect to the birth parameter. More precisely, the two functions of $\beta$ are nondecreasing:

$$
\begin{aligned} \theta_{W}(\beta) &=P_{\beta}\left(\xi_{t} \neq \varnothing \text { for all } t | \xi_{0}=\{0\}\right) \\ \theta_{S}(\beta) &=P_{\beta}\left(0 \in \xi_{t} \text { infinitely often } | \xi_{0}=\{0\}\right) \end{aligned}
$$
again, $\xi_t$ is a collection of occupied sites. Now, the critical values are defined as:

$$
\beta_{W}=\sup \left\{\beta: \theta_{W}(\beta)=0\right\} \quad \text { and } \quad \beta_{S}=\sup \left\{\beta: \theta_{S}(\beta)=0\right\}
$$

### Lemma for attractiveness
For all $(x, t) \in \mathbb{Z}^{d} \times \mathbb{R}_{+}$

$$
P\left(x \in \xi_{t}^{1}\right) \leq P\left(x \in \xi_{t}^{2}\right) \quad \text { when } \quad \xi_{0}^{1} \subset \xi_{0}^{2} \quad \text { and } \quad \beta_{1}=\beta_{2}
$$

### Definition of the upper invariant measure

Letting $\mu_{t}$ be the distribution at time $t$ of the contact process starting from the all-occupied configuration, it is clear that $\mu_{t} \leq \mu_{0} .$ Then, using attractiveness and the fact that the contact process is a Markov process, we deduce that

$$
\mu_{t+s} \leq \mu_{s} \quad \text { for all } \quad s \geq 0
$$
In particular, starting from the all-occupied configuration,

$$
\bar{\mu}=\lim _{t \rightarrow \infty} \mu_{t}
$$

This is the upper invariant measure of the contact process and, this is the **largest possible stationary distribution**.

## Self-duality
The dual process is defined from a graphical representation of the process. **When we assume it is fixed,** we have the following definition:

### Definition of the dual paths
We say that there is a path $(z, s) \rightarrow(x, t)$ if there exist
$$
s=s_{0}<s_{1}<\cdots<s_{n+1}=t \quad \text { and } \quad z=x_{0}, x_{1}, x_{2}, \ldots, x_{n}=x \in \mathbb{Z}^{d}
$$
such that the following two conditions hold:

- for $i=1,2, \ldots, n,$ there is an arrow $x_{i-1} \rightarrow x_{i}$ at time $s_{i}$ and

- for $i=0,1, \ldots, n,$ there is no $\times$ (i.e., death) on the segment $\left\{x_{i}\right\} \times\left(s_{i}, s_{i+1}\right)$

If this holds, we also say that there is a dual path $(x, t) \rightarrow(z, s)$.

In other words, **dual paths are defined as paths except that they evolve backward in time and follow the arrows of the graphical representation in the opposite direction**.

### Definition of the dual process starting at (x,t)
$$
\hat{\xi}_{s}(x, t)=\left\{z \in \mathbb{Z}^{d}: \text { there is a dual path }(x, t) \rightarrow(z, t-s)\right\}
$$
for all $0 \leq s \leq t$

### Properties of the dual process

1. the distribution of the process starting from a single individual at vertex x is equal to the distribution of its dual process starting at (x, t),i.e.,

$$
\left\{\xi_{s}: 0 \leq s \leq t\right\}\overset{d}{=}\left\{\hat{\xi}_{s}(x, t): 0 \leq s \leq t\right\}
$$

2. From the equality above, we have another property of **duality relationship**:

$ x \in \xi_{t}$ if and only if: $\xi_{S} \cap \hat{\xi}_{t-s}(x, t) \neq \varnothing$

For the contact process starting from the all-occupied configuration,
$$
P\left(x \in \xi_{t}\right)=P\left(\hat{\xi}_{t}(x, t) \cap \xi_{0} \neq \varnothing\right)=P\left(\hat{\xi}_{t}(x, t) \neq \varnothing\right)
$$
and the upper invariant measure becomes:
$$
\begin{aligned} \bar{\mu}(\{x \in \xi\}) &=\lim _{t \rightarrow \infty} P\left(x \in \xi_{t}\right) \\ &=\lim _{t \rightarrow \infty} P\left(\hat{\xi}_{t}(x, t) \neq \varnothing\right)=\lim _{t \rightarrow \infty} P_{x}\left(\xi_{t} \neq \varnothing\right) \\ &=P_{x}\left(\xi_{t} \neq \varnothing \text { for all } t\right)=\theta_{W}(\beta) \end{aligned}
$$
In other words, self-duality implies that the probability that a given vertex is occupied under the upper invariant measure is equal to the probability that the process starting with a single individual survives weakly.

## The critical value
The strong survival implies weak survival, we clearly have $\beta_{W} \leq \beta_{S} .$ In fact, it is known that, at least for the contact process on the infinite integer lattice, these two critical values are equal.

### Theorem for critical values
For the process on $\mathbb{Z}^{d},$ we have $\beta_{c} \geq 2 d /(2 d-1)>1$

### Complete convergence theorem
**If the contact process survives, then it must converge to its
upper invariant measure.**

### Theorem for extinction time at subcritical case
Assume that $\beta<\beta_{c} .$ Then,
$$
P\left(\xi_{t} \neq \varnothing | \xi_{0}=\{0\}\right) \leq \exp (-\alpha t) \quad \text { for some } \quad \alpha>0
$$
Again, $\xi_{0}=\{0\}$ means single individual at the origin at the beginning. In other words, **the process dies out exponentially fast.**

## Single-type Swap Process

The algorithm introduced here is inspired by my previous paper on the atomic stacking in porous lattice structure. You can refer to the paper [here](https://pubs.acs.org/doi/10.1021/acs.jpcc.8b12455#) for more information. In the system described in the paper, we have each site either empty or occupied by a cation. Each cation can interact with its neighboring empty sites at a rate of $\beta$, or stays at the same site at rate 1. On a 2D space, the interaction rate at each occupied site $\beta_i$ is defined as:
$$
\beta_i = \sum_{\hat j}\xi_t(\hat j)
$$
where $\xi_t(\hat j)$ is the occupancy of each neighboring site at time $t$. $\xi_t(\hat j)=1$ when the site is occupied. Note that $\beta_i=0$ when all the neighbors of atom $i$ are occupied. Therefore, we construct the following rules for this swap process:

1. Attach a Poisson process to each cation, and the time $\tau$ lapsed before next event happens on cation $i$ follows:
$$
\tau=min\{X_i:X_i\sim exponential(\beta_i+1)\}
$$

2. For cation $i$, it has a probability of $1/(\beta_i+1)$ to stay and $\beta_i/(\beta_i+1)$ to swap.

3. Uniformly choose a neighbor around the cation $i$, and only swap the location of $i$ with its neighbor when the neighbor is empty.

4. repeat 1-3 until the time reaches a prefixed value, T.

```python
import numpy as np
from scipy.stats import expon
from numpy.random import choice
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]

def calcBeta(u):
    d1,d2=u.shape
    betaMtx = np.zeros((d1,d2))
    for x in range(d1):
        for y in range(d2):
            if u[x][y]>0: # we only have beta for occupied sites
                betaMtx[x][y]=u[(x+d1+1)%d1][y]+u[(x+d1-1)%d1][y]+u[x][(y+d2-1)%d2]+u[x][(y+d2+1)%d2]
    return betaMtx.ravel()

def singleSwap(n, n1, T):
    # We start with random initial configuration
    N = n**2
    universe = np.zeros(n**2,dtype=int)
    universe[choice(N,size=n1,replace=False)]=1
    universe = universe.reshape((n,n))
    inituniv=universe
    bmtx = calcBeta(universe)
    tlst=[0]
    guardlst = [np.mean(bmtx)/4] # calculate the guard number defined in the paper
    nbs = [[-1,0],[1,0],[0,1],[0,-1]]
    while tlst[-1]<T:
        curx,cury, curi=0,0,0
        first = T
        for i in range(N):
            if bmtx[i]>0: # only the cations with occupied neighbors will swap
                temp = expon(scale=1/bmtx[i]).rvs()
                if temp<first:
                    first = temp
                    curx,cury=np.unravel_index(i,(n,n))
                    curi = i
        tlst.append(tlst[-1]+first)
        b=choice(4)
        if universe[(curx+n+nbs[b][0])%n][(cury+n+nbs[b][1])%n]==0:
            if choice([0,1],p=[1/(bmtx[curi]+1), bmtx[curi]/(bmtx[curi]+1)]):
                universe[(curx+n+nbs[b][0])%n][(cury+n+nbs[b][1])%n]=1
                universe[curx][cury]=0
                
        bmtx = calcBeta(universe)
        guardlst.append(np.mean(bmtx)/4)
    return universe,inituniv, tlst, guardlst
```

```python
u,initu,Ts,Gs=singleSwap(6,18,10)
plt.imshow(initu)
plt.colorbar()
```

**The initial configuration above shows no pattern for the occupied sites distribution.**

```python
plt.imshow(u)
plt.colorbar()
```

**Compare the final configuration with the initial configuration, we find that the chessboard pattern is formed is some region after some time. Now, let's check the guard number.**

```python
plt.step(Ts,Gs)
plt.xlabel('Time')
plt.ylabel('Guard number')
```

**Looks like the guard number is fluctuating around 0.18, let's run the process for longer time**
