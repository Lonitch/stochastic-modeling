## Overview

In percolation models, the actual spatial distance between individuals is no longer ignored and is in fact the key to constructing the models. The processes, however, no longer evolve in time, though, as we will see in a moment, one can interpret each realization as the final set resulting from the invasion of a rumor or disease that spreads in space.

More precisely, percolation models are random graphs and therefore static random structures. These processes are generally difficult to study because there are only a few universal techniques available and the analysis of a new model often requires the development of new mathematical tools.

## Bond Percolation in Two Dimension

Imagine that each site of the two-dimensional integer lattice is occupied by an individual and that, independently of each other, any two nearest neighbors are friends with probability $p$. If an infection can only spread from friend to friend, *what is the probability that an infection introduced at the origin will result in an epidemic, meaning that
infinitely many individuals get infected?*

Bond percolation in two dimensions is simply the process that consists of the countable collection of independent Bernoulli random variables with success probability $p$ that keep track of the friendship relationships. More precisely, let
$$
\xi(e) \sim \text { Bernoulli }(p) \quad \text { for each } \quad e \in E=\left\{(x, y) \in \mathbb{Z}^{2} \times \mathbb{Z}^{2}:\|x-y\|=1\right\}
$$
In the terminology of percolation theory, edge $e$ is said to be
$$
\text { open when } \xi(e)=1 \text { and closed when } \xi(e)=0
$$
We say that there is an open path between $x$ and $y,$ which we write $x \leftrightarrow y,$ whenever there exists a sequence of vertices $x=x_{0}, x_{1}, \ldots, x_{n}=y$ such that
$$
\left(x_{i}, x_{i+1}\right) \in E \quad \text { is open } \quad \text { for all } i=0,1, \ldots, n-1
$$
Define the **open cluster** at vertex $x$ as
$$
C_{x}=\left\{y \in \mathbb{Z}^{2}: x \leftrightarrow y\right\}
$$
The cluster $C_0$ is the connected component of this subgraph **containing the origin**. In particular, the event that an epidemic occurs, is the event that **this cluster is infinite**. Thus, the probability of an epidemic is given by the so-called **percolation probability**
$$
\theta(p)=P_{p}\left(\left|C_{0}\right|=\infty\right) \quad \text { where } \quad\left|C_{0}\right|=\operatorname{card}\left(C_{0}\right)
$$
where $card(C_0)$ is the *cardinal number of the set* $C_0$.

## Monotonicity of the percolation propability

It should be intuitively clear that **the percolation probability is a nondecreasing function of the density $p$ of open edges**. This can be made rigorous using a so-called **coupling argument.** The basic idea is to find a joint construction of two random variables or stochastic processes, which is called a **coupling**, such that one dominates the other one with probability one.

### Theorem/proof for the monotonicity 
**The function $p \mapsto \theta(p)$ is nondecreasing.**

*proof:* Having two parameters $p_{1} \leq p_{2},$ the objective is to find a joint construction of bond percolation with parameter $p_{1}$ and bond percolation with parameter $p_{2}$ such that the latter dominates the former with probability one. To do this, we let
$$
U(e) \sim \text { Uniform }(0,1) \text { for each edge } e \in E
$$
be independent and define, for $i = 1, 2$, the two collections
$$
\xi_{i}(e)=1 \text { if and only if } U(e) \leq p_{i} \text { for each edge } e \in E
$$
These random variables have the following three properties:

- The random variables $\xi_{i}(e), e \in E,$ are independent
- $P\left(\xi_{i}(e)=1\right)=P\left(U(e) \leq p_{i}\right)=p_{i}$ hence $\left\{\xi_{i}(e)|: e \in E\right\}$ indeed defines a realization of a bond percolation process with parameter $p_i$
- the two percolation processes are not independent and we have
$$
\begin{aligned}\left\{\xi_{1}(e)=1\right\} &=\left\{U(e) \leq p_{1}\right\} \\ & \subset\left\{U(e) \leq p_{2}\right\}=\left\{\xi_{2}(e)=1\right\} \quad \text { for each } e \in E \end{aligned}
$$
In particular, we have $\left(\xi_{1}\right) \leq\left(\xi_{2}\right),$ i.e., each of the coordinates of $\xi_{1}$ is, with probability one, at most equal to its counterpart for $\xi_{2} .$
Now, define
$$
f(\xi)=\mathbf{1}\left\{\left|C_{0}(\xi)\right|=\infty\right\} \quad \text { for all } \quad \xi: E \rightarrow\{0,1\}
$$
where $C_{0}(\xi)$ is the cluster at the origin. since $\left(\xi_{1}\right) \leq\left(\xi_{2}\right)$ and since $f$ is clearly nondecreasing, we conclude that
$$
\theta\left(p_{1}\right)=P\left(\left|C_{0}\left(\xi_{1}\right)\right|=\infty\right)=E\left(f\left(\xi_{1}\right)\right) \leq E\left(f\left(\xi_{2}\right)\right)=\theta\left(p_{2}\right)
$$

## The critical phenomenon
Here, we introduce a critical value:
$$
p_{c}=\sup \{p: \theta(p)=0\}
$$
Above this critical value, there is a positive probability of an epidemic, while below this critical value the probability of an epidemic is equal to zero.

Below the critical value, *the size of a typical cluster decays exponentially, whereas above the critical value there exists exactly one infinite cluster of open edges*. More precisely, we have the following two results.
### Theorem for critical values
**First theorem:** Assume that $p<p_{c}$. Then,
$$
P\left(\left|C_{0}\right|>n\right) \leq \exp (-\alpha n) \text { for some } \alpha>0
$$
Assume that $p>p_{c} .$ Then, the subgraph induced by the open edges has a unique infinite connected component. This infinite connected component is called the *infinite percolation cluster*.

*Critical values for spatially explicit models such as percolation processes and interacting particle systems are usually impossible to compute.*

**Second theorem:** $p_c$ can be proved as $1/2$ for bond percolation models by using planar graph and planar duality:

- **Planar graph**: A planar graph is a graph that can be drawn on the plane in such a way that its edges do not intersect each other.

- **Planar duality:** The dual graph of a planar graph G is the graph

    - that has a vertex corresponding to each plane region of G
    - and an edge joining two neighboring regions for each edge in G.
    
**Third theorem**: In d dimensions, we have 
$$
1 /(2 d-1) \leq p_{c} \leq 1 / 2
$$

## Oriented site percolation in two dimensions

Let $\mathscr{G}$ be the set of sites
$$
\mathscr{G}=\left\{(x, n) \in \mathbb{Z} \times \mathbb{Z}_{+}: x+n \text { is even }\right\}
$$
which we turn into a directed graph by putting an arrow
$$
(x, n) \rightarrow\left(x^{\prime}, n^{\prime}\right) \quad \text { if and only if } \quad x^{\prime}=x \pm 1 \text { and } n^{\prime}=n+1
$$
Site percolation consists of a collection of independent Bernoulli random variables with the same success probability $p$ indexed by the set $\mathscr{G},$ i.e.,
$$
\xi(x, n) \sim \text { Bernoulli }(p) \quad \text { for each site } \quad(x, n) \in \mathscr{G}
$$
Site (x,n) is said to be
$$
\text { open when } \xi(x, n)=1 \text { and closed when } \xi(x, n)=0
$$
If we think of the orientation of the graph as gravity, we can define the set of **wet sites** at level $n$ as
$$
W_{n}=\{y \in \mathbb{Z}:(x, 0) \rightarrow(y, n) \text { for some } x \in 2 \mathbb{Z}\}
$$
*Note that wet sites are open, but open sites are not necessarily wet.* Using the terminology of epidemiology, one can think of open as susceptible and wet as infected: susceptible individuals become infected only if they can be reached by a path of infection.

### Theorem for the critical value
$$
\text { We have } p_{c} \leq 1-3^{-4}=80 / 81
$$

**Enough for the theory, let's first implement bond percolation model**

```python
import numpy as np
from scipy.stats import bernoulli
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6, 6]

# to generate a realization of bond or site percolation, 
# we only need to create a collection of Bernoulli random 
# variables to determine whether the edges or vertices 
# are open or closed.

def bond_perco(n, p):
    # n: size of the 2D universe
    # p: probability that a edge is open
    rv = bernoulli(p)
    edges = rv.rvs(size=2*n*(n-1))
    curr = 0
    for i in range(n):
        for j in range(n-1):
            if edges[curr]==1:
                plt.plot([i,i],[j,j+1],'k*-')
            else:
                plt.plot([i,i],[j,j+1],'k*')
            curr+=1
                
    for j in range(n):
        for i in range(n-1):
            if edges[curr]==1:
                plt.plot([i,i+1],[j,j],'k*-')
            else:
                plt.plot([i,i+1],[j,j],'k*')
            curr+=1
```

### Implementation of Oriented site percolation

Because the graph is now directed, after the configuration of open and closed sites has been determined, we also identify the wet sites recursively starting from level zero. The following program again displays the spatial configuration in the form of an array of characters where open sites are circles `o` when there are wet and stars `*` when they are not wet, while we use an `unfilled triangle` for the closed sites.

```python
import numpy as np
from scipy.stats import bernoulli
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6, 6]

def oriented_perco(m, n,p):
    rv = bernoulli(p)
    sites = np.zeros((m,n),dtype=int)
    wet = np.zeros((m,n),dtype=int)
    sites[0][::2]=1
    wet[0][::2]=1
    for i in range(1,m):
        for j in range(n):
            if (i+j)%2==0 and rv.rvs():
                sites[i][j] = 1
    for i in range(1,m):
        for j in range(n):
            if (i+j)%2==0:
                if 0<j<n-1:
                    if sites[i][j]==1:
                        if wet[i-1][j+1]==1 or wet[i-1][j-1]==1:
                            wet[i][j]=1
                elif j==n-1:
                    if sites[i][j]==1 and wet[i-1][j-1]==1:
                        wet[i][j]=1
                elif j==0:
                    if sites[i][j]==1 and wet[i-1][j+1]==1:
                        wet[i][j]=1
                
    return sites, wet

def plot_oriented_perco(site, wet):
    m,n = site.shape
    for i in range(m):
        for j in range(n):
            if site[i][j]==0:
                plt.plot(j,m-i,'k^',markerfacecolor='None')
            elif site[i][j]==1 and wet[i][j]==1:
                plt.plot(j,m-i,'ko',markerfacecolor='None')
            else:
                plt.plot(j,m-i,'k*')
    for i in range(1,m):
        for j in range(n):
            if j==0:
                if wet[i-1][j+1]==1 and wet[i][j]==1:
                    plt.arrow(j+1, m-i+1,-1,-1,color='k',width=0.03)
            elif j==n-1:
                if wet[i-1][j-1]==1 and wet[i][j]==1:
                    plt.arrow(j-1, m-i+1,+1,-1,color='k',width=0.03)
            else:
                if wet[i-1][j+1]==1 and wet[i][j]==1:
                    plt.arrow(j+1, m-i+1,-1,-1,color='k',width=0.03)
                if wet[i-1][j-1]==1 and wet[i][j]==1:
                    plt.arrow(j-1, m-i+1,+1,-1,color='k',width=0.03)
```

```python
S,W=oriented_perco(8,8,0.55)
plot_oriented_perco(S,W)
```

```python
S,W=oriented_perco(8,8,0.8)
plot_oriented_perco(S,W)
```

```python
S,W=oriented_perco(8,8,0.3)
plot_oriented_perco(S,W)
```
